```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
df = pd.read_csv("./data/penguins.csv")
```

```python
#see head of the dataframe
df.head()
```

```python
df.groupby(['sex'])['sex'].count()
```

```python
m = df.groupby(['sex'])['sex'].count()
```

```python
df.groupby(['species','island','sex'])['sex'].count()
```

```python
ax = df.groupby(['species','island','sex'])['sex'].count().plot(kind='bar',color=['pink', 'lightblue']);
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2,
            y + height + 0.01,
            '{:.0%}'.format(height / m.sum()),
            ha='center',
            va='bottom')

plt.title("Male/Female  distribution for each and island and species")
```
